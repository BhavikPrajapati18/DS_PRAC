Feature Scaling using Standardization and Normalization

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Dataset
data = {
    'Country': ['France', 'Spain', 'Germany', 'Spain', 'Germany', 'France', 'Spain', 'France', 'Germany', 'France'],
    'Age': [44, 27, 30, 38, 40, 35, 52, 48, 50, 37],
    'Salary': [72000, 48000, 54000, 61000, 80500, 58000, 52000, 79000, 83000, 67000],
    'Purchased': ['No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)

# Select only numerical features
num_data = df[['Age', 'Salary']]

# Standardization (Z-score)
standard_scaler = StandardScaler()
standardized = standard_scaler.fit_transform(num_data)

# Normalization (Min-Max)
minmax_scaler = MinMaxScaler()
normalized = minmax_scaler.fit_transform(num_data)

# Create DataFrames for comparison
standardized_df = pd.DataFrame(standardized, columns=['Age_Standardized', 'Salary_Standardized'])
normalized_df = pd.DataFrame(normalized, columns=['Age_Normalized', 'Salary_Normalized'])

# Merge all results
result = pd.concat([df, standardized_df, normalized_df], axis=1)
print(result)
```

 Principal Component Analysis (PCA) on Iris Dataset

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardize features before PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create DataFrame
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['Target'] = y

# Visualize
plt.figure(figsize=(8, 6))
for label in set(y):
    plt.scatter(
        pca_df[pca_df['Target'] == label]['PC1'],
        pca_df[pca_df['Target'] == label]['PC2'],
        label=iris.target_names[label]
    )

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on Iris Dataset')
plt.legend()
plt.grid(True)
plt.show()
```
